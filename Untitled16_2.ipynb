```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import datasets, transforms
import copy
```

```python
class MultiViewTransform:
  def __init__(self):
    self.global_transform = transforms.Compose([
        transforms.RandomResizedCrop(64, scale=(0.5, 1.0)),
        transforms.RandomHorizontalFlip(),
        transforms.ColorJitter(0.4,0.4,0.4,0.1),
        transforms.RandomGrayscale(p=0.2),
        transforms.GaussianBlur(3),
        transforms.ToTensor()
        ])

    self.local_transform = transforms.Compose([
        transforms.RandomResizedCrop(64, scale=(0.2, 0.5)),
        transforms.RandomHorizontalFlip(),
        transforms.ColorJitter(0.4,0.4,0.4,0.1),
        transforms.ToTensor()
        ])

  def __call__(self, x):
    views = [self.global_transform(x), self.global_transform(x)]
    views += [self.local_transform(x) for _ in range(2)]

    return views
```

```python
# Load in the data
transform = MultiViewTransform()

dataset = datasets.CIFAR10(root="./data", train=True, download=True, transform=transform)

loader = torch.utils.data.DataLoader(
    dataset,
    batch_size=32,
    shuffle=True,
    num_workers=2,
    pin_memory=True

)
```

```python
class PatchEmbed(nn.Module):
  def __init__(self, dim=192):
    super().__init__()
    self.proj = nn.Conv2d(3, dim, kernel_size=8, stride=8)

  def forward(self, x):
    x = self.proj(x)
    x = x.flatten(2).transpose(1,2)
    return x

class Attention(nn.Module):
  def __init__(self, dim, heads=3):
    super().__init__()
    self.heads = heads
    self.scale = dim ** -0.5
    self.qkv = nn.Linear(dim, dim*3)
    self.proj = nn.Linear(dim, dim)

  def forward(self, x):
    B,N,C = x.shape
    qkv = self.qkv(x).reshape(B,N,3,self.heads,C//self.heads)
    q,k,v = qkv.permute(2,0,3,1,4)

    attn = (q @ k.transpose(-2,-1)) * self.scale
    attn = attn.softmax(dim=-1)

    x = (attn @ v).transpose(1,2).reshape(B,N,C)
    return self.proj(x)

class Block(nn.Module):
  def __init__(self, dim):
    super().__init__()
    self.norm1 = nn.LayerNorm(dim)
    self.attn = Attention(dim)
    self.norm2 = nn.LayerNorm(dim)
    self.mlp = nn.Sequential(
        nn.Linear(dim, dim*4),
        nn.GELU(),
        nn.Linear(dim*4, dim)
        )

  def forward(self, x):
    x = x + self.attn(self.norm1(x))
    x = x + self.mlp(self.norm2(x))
    return x

class MiniViT(nn.Module):
  def __init__(self, dim=192, depth=6):
    super().__init__()
    self.patch = PatchEmbed(dim)
    self.cls = nn.Parameter(torch.randn(1,1,dim))
    self.pos = nn.Parameter(torch.randn(1,65,dim))
    self.blocks = nn.Sequential(*[Block(dim) for _ in range(depth)])
    self.norm = nn.LayerNorm(dim)

  def forward(self, x):
    x = self.patch(x)
    B,N,_ = x.shape
    cls = self.cls.expand(B,-1,-1)
    x = torch.cat([cls,x],dim=1)
    x = x + self.pos[:,:N+1]
    x = self.blocks(x)
    x = self.norm(x)
    return x[:,0]
```

```python
class DINOHead(nn.Module):
  def __init__(self, in_dim=192, out_dim=128):
    super().__init__()
    self.mlp = nn.Sequential(
        nn.Linear(in_dim, 256),
        nn.GELU(),
        nn.Linear(256, out_dim)
        )

  def forward(self, x):
    return self.mlp(x)
```

```python
student = nn.Sequential(MiniViT(), DINOHead())
teacher = copy.deepcopy(student)

for p in teacher.parameters():
  p.requires_grad = False
```

```python
import torch.nn.functional as F
class DINOLoss(nn.Module):
  def __init__(self, out_dim=128, momentum=0.9):
    super().__init__()
    self.register_buffer("center", torch.zeros(1, out_dim))
    self.momentum = momentum

  def forward(self, student_out, teacher_out):

    # student
    student_out = student_out / 0.1
    student_logprob = F.log_softmax(student_out, dim=-1)

    # teacher (detach + stabilize)
    teacher_out = teacher_out.detach()

    # subtract center BEFORE scaling
    teacher_centered = teacher_out - self.center

    # clamp to avoid overflow
    teacher_centered = torch.clamp(teacher_centered, -10, 10)
    teacher_prob = F.softmax(teacher_centered / 0.04, dim=-1)
    loss = -(teacher_prob * student_logprob).sum(dim=-1).mean()

    # update center using RAW teacher output (CRITICAL FIX)
    with torch.no_grad():
      batch_center = teacher_out.mean(dim=0, keepdim=True)
      self.center = self.center * self.momentum + (1 - self.momentum) * batch_center
      return loss
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

student.to(device)
teacher.to(device)

torch.backends.cudnn.benchmark = True

criterion = DINOLoss().to(device)
optimizer = torch.optim.Adam(student.parameters(), lr=5e-5)
loss_history = []

for epoch in range(15):
    epoch_loss = 0
    n_batches = 0

    for views, _ in loader:

        views = [v.to(device) for v in views]

        teacher_out = [teacher(v).detach() for v in views[:2]]
        student_out = [student(v) for v in views]

        loss = 0
        count = 0

        for i, t in enumerate(teacher_out):
            for j, s in enumerate(student_out):
                if j == i:
                    continue
                loss += criterion(s, t)
                count += 1

        loss /= count


        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(student.parameters(), 1.0)
        optimizer.step()

        with torch.no_grad():
            for ps, pt in zip(student.parameters(), teacher.parameters()):
                pt.data = 0.996 * pt.data + 0.004 * ps.data

        epoch_loss += loss.item()
        n_batches += 1

    avg_loss = epoch_loss / n_batches
    loss_history.append(avg_loss)

    print(f"Epoch {epoch}: {avg_loss:.4f}")
```
